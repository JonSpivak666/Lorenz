# Sistema de Lorenz y el Caos

El sistema de Lorenz es un sistema de ecuaciones diferenciales ordinarias tridimensionales, desarrollado por Edward N. Lorenz en 1963, para modelar las convecciones atmosféricas. Estas ecuaciones son conocidas por su comportamiento caótico, lo que implica una sensibilidad extrema a las condiciones iniciales.


## Ecuaciones del Sistema de Lorenz

Las ecuaciones del sistema de Lorenz son:


\begin{cases}
\dot{x} = \sigma (y - x) \\
\dot{y} = x (r - z) - y \\
\dot{z} = xy - bz
\end{cases}


donde $\sigma$, $r$, y $b$ son parámetros positivos que representan el número de Prandtl, el número de Rayleigh y una razón geométrica del sistema, respectivamente. En este ejemplo usaremos los valores comúnmente usados para estos parámetros son $\sigma = 10$, $r = 28$, y $b = \frac{8}{3}$.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def lorenz(t, X, sigma=10, beta=8/3, rho=28): #Definimos el Sistema
    x, y, z = X
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]
```

```python
# Condiciones iniciales y tiempo
X0 = [1.0, 1.0, 1.0]
t_span = (0, 50)
t_eval = np.linspace(*t_span, 10000)

#Sol numerica
sol = solve_ivp(lorenz, t_span, X0, t_eval=t_eval)
```

```python
fig = plt.figure() #Grafica correspondiente
ax = fig.add_subplot(111, projection='3d')
ax.plot(sol.y[0], sol.y[1], sol.y[2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Atractor de Lorenz')
plt.show()
```
